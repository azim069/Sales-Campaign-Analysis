##FILL "duration" and clean the duration data

```python
new_df[['duration']]
```

```python
new_df.groupby("duration")['duration'].count()
```

```python
# Function to replace specific substrings with 0
def replace_with_0(value):

    if value is None:
        return None  # Return None for None values
    substrings_to_replace = ['Amanat Khan Bari BZ’', '$ days', '$400','2 Ad','2nd','35$', '20$','60$', 'Post: "মার্কেটপ্লেসের বাইরে SEO ক্লায়েন্ট পাওয়ার..."']
    for substring in substrings_to_replace:
        if substring in value:
            return 0
    return value
# Map the function to the 'duration' column
new_df['duration(days)'] = new_df['duration'].map(replace_with_0)
new_df[['duration(days)']]
```

```python
new_df[['duration(days)','content']]
```

```python
# Define a regular expression pattern to extract numeric values followed by "Days"
new_df['duration(days)'] = new_df['duration(days)'].str.replace(' days', '').str.replace(' Days', '').str.replace('Days', '').str.replace('days', '')
new_df[['duration(days)']]
```

```python
new_df['duration(days)'] = pd.to_numeric(new_df['duration(days)'], errors='coerce').fillna(0).astype(int)
```
